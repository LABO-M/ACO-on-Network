```julia
using CSV
using DataFrames

# パラメータの設定
T = 100
r = 3
w_list = [10, 1.0, 0, -0.99, -1.0]

for w in w_list
    # ネットワークの生成
    _, _, link_matrix = Network.generate_network(T, r, w)

    # マトリックスをDataFrameに変換
    df = DataFrame(link_matrix, :auto)

    # ファイル名の設定
    w_str = replace(string(w), ".0" => "")
    filename = "data/link/T100_omega$(w_str).csv"

    # CSVファイルに保存
    CSV.write(filename, df)
end
```
