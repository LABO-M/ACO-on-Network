```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
omega10 = pd.read_csv('data/k_out/T1M_omega10.csv')
omega1 = pd.read_csv('data/k_out/T1M_omega1.csv')
omega0 = pd.read_csv('data/k_out/T1M_omega0.csv')
omega_099 = pd.read_csv('data/k_out/T1M_omega-0.99.csv')
omega_1 = pd.read_csv('data/k_out/T1M_omega-1.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 
def plot_scatter(df, title):
    # k_out
    frequency = df['k_out'].value_counts().sort_index()
    # 
    probability = frequency / frequency.sum()

    # 
    plt.figure(figsize=(6, 7))
    plt.scatter(probability.index, probability.values, color='cornflowerblue')
    plt.title(title, fontsize=24)
    plt.xlabel(r'$k_{out}$', fontsize=22)
    plt.ylabel('Probability', fontsize=18)
    # plt.ylim(0, 0.51)
    plt.xticks(fontsize=18)
    plt.yticks(fontsize=18)
    plt.xscale('log')
    plt.yscale('log')
    plt.subplots_adjust(left=0.17, bottom=0.15, right=0.95, top=0.85)
    # df
    file_name = title.replace('$\omega=', 'omega').replace('.0$', '').replace('.', '').replace('$', '')
    plt.savefig('fig/' + file_name + '.eps', format='eps', dpi=300)
    plt.show()

# 
plot_scatter(omega10, r'$\omega=10.0$')
plot_scatter(omega1, r'$\omega=1.0$')
plot_scatter(omega0, r'$\omega=0.0$')
plot_scatter(omega_099, r'$\omega=-0.99$')
plot_scatter(omega_1, r'$\omega=-1.0$')
```
