```python
import pandas as pd
import numpy as np

w_1 = pd.read_csv("data/Zt/N100_T1M_r100_omega-1.0_alpha1.01.csv")
w_099 = pd.read_csv("data/Zt/N100_T1M_r100_omega-0.99_alpha1.01.csv")
```

```python
w_1.head()
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(7, 9))

plt.fill_between(w_1['t'], w_1['Z']-w_1['std'], w_1['Z']+w_1['std'], color='k', alpha=0.2)
plt.plot(w_1['t'], w_1['Z'], '0.5', linestyle=':', label=r'$\omega=-1.0, \alpha=1.01$', color='cornflowerblue', linewidth=3)

plt.plot(w_099['t'], w_099['Z'], '0.5', linestyle=':', label=r'$\omega=-0.99, \alpha=1.01$', color='lightcoral', linewidth=3)
plt.fill_between(w_099['t'], w_099['Z']-w_099['std'], w_099['Z']+w_099['std'], color='k', alpha=0.2, label='standard deviation')


# 軸とタイトル
plt.xlabel('$t$', fontsize=30)
plt.ylabel('$Z(t)$', fontsize=30)
# plt.title('Semi-log plot of df', fontsize=16)
plt.legend(fontsize=22, loc='upper left')

# y軸の指数表示をオフ
plt.gca().yaxis.set_major_formatter(plt.ScalarFormatter())
plt.xlim(1e2, 1e6)
plt.ylim(0.48, 1.01)

plt.grid(True, which="both", ls="--")
plt.xscale('log')

plt.xticks(fontsize=24)
plt.yticks(fontsize=24, rotation=90, va='center')

plt.show()
```
